## 데이터프레임 정렬

```python
import numpy as np
import pandas as pd
print("DataFrame: ")
data = {
    'col1':[2,1,9,8,7,4],
    'col2':['A','A','B',np.nan,'D','C'],
    'col3':[0,1,9,4,2,3]
}
index = [1,3,2,5,0,4]
df = pd.DataFrame(data,index=index)
df
```

```python
# sort_index axis=0(행) 을 축으로 정렬
df = df.sort_index(axis=0)
df
```

```python
# sort_index axis=1(열) 을 축으로 정렬, 내림차순
df = df.sort_index(axis=1, ascending=False)
df
```

```python
# sort_value 1
sorted_df1 = df.sort_values('col1', ascending=True)
sorted_df1
```

```python
# sort_value 2
sorted_df2 = df.sort_values(['col2','col1'],ascending=[True,False])
sorted_df2
```

## 데이터 분석용 함수

```python
# 데이터 프레임의 index와 columns
import numpy as np
import pandas as pd
data = {
    'korean': [50,60,70,30],
    'math':[30,np.nan,40,20]
}
index = ['a','b','c','d']
df = pd.DataFrame(data,index=index)
print(df.index)
print(df.columns)
df
```

```python
# count() 함수
print(df.count(axis=0)) # axis = 0 기본값(생략가능)
print(df.count(axis=1))
```

```python
# max(), min(), sum(), mean()
print(df.max())
print(df.max(axis=1))
print(df.min())
print(df.min(axis=1))
print(df.sum())
print(df.sum(axis=1))
print(df.mean())
print(df.mean(axis=1))
print(df.mean(axis=1,skipna=False)) # NaN값 보이게 하기
```

```python
# 결측값 대체 fillna
avg = df['math'].mean()
df['math']=df['math'].fillna(avg) 
df
```

## 그룹으로 묶기 (groupby)

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'key'   : ['A','A','A','A','B','C'],
    'data1':[1,2,3,1,2,3,],
    'data2':[4,4,6,0,6,1],
    # 'data3':['안녕','a','b','하세요','pple','3']
    })
print("DataFrame:")
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby(['key','data1']).sum()#.reset_index()
```

```python
df.groupby('key').aggregate(['min','median','max'])
```

```python
df.groupby('key').aggregate({'data1':'max','data2':'sum'})
```

```python
# filter 메서드
t = list(range(1,11))
result = []
def is_even(x):
    return True if x%2 == 0 else False

for i in t:
    if is_even(i):
        result.append(i)
print(result)
```

```python
df.groupby('key').mean()
```

```python
def filter_mean(x):
    return x['data2'].mean() > 3
df.groupby('key').filter(filter_mean)
```

```python
df1 = df.groupby('key').get_group('A')
df2 = df.groupby('key').get_group('B')
df3 = df.groupby('key').get_group('C')
df4 = pd.concat([df1,df3], axis=1)
df5 = pd.concat([df2,df3], axis=1)
# 평균값으로 채운다.
#means = df2['data2'].astype(int).mean()
#means
#df5.fillna(method='ffill')

# 전후의 값으로 NaN값을 채운다.
#df5.ffill()
#df5.bfill()
```

```python
import pandas as pd
df = pd.DataFrame([
    ['20210901','A',1],
    ['20210901','B',10],
    ['20210901','C',100],
    ['20210902','A',2],
    ['20210902','B',20],
    ['20210902','C',200],
    ['20210903','A',3],
    ['20210903','B',30],
    ['20210903','C',300],
], columns = ['date','type','volume'])
df
```

```python
# group 세로형 데이터프레임 유지
# pivot, pivot_table 가로형 재결합, 재가공의 느낌
pivot1 = df.pivot(index="date",columns="type",values="volume")
pivot1
```

```python
pivot1.columns = pivot1.columns.values
pivot1
```

```python
pivot1.reset_index(level=0,inplace=True)
pivot1
```

```python
pivot1 = pivot1.set_index('date')
pivot1
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2,figsize=(13,5))
pivot1.plot(kind='bar',ax=ax[0])
pivot1.plot(kind='barh',ax=ax[1])
```

```python
# í…ŒìŠ¤íŠ¸ 9 ì€í–‰ ì˜ˆê¸ˆ - ìº¡ìŠí™” o, í´ë˜ìŠ¤ ë³€ìˆ˜
class BankAccount:
    
    interest_rate = 0.02 #í´ë˜ìŠ¤ ë³€ìˆ˜, ì´ììœ¨, í•¨ìˆ˜ 1ê°œê°€ ì•„ë‹Œ ëª¨ë“  í•¨ìˆ˜ì— ì ìš©ì´ ë˜ëŠ” ê²ƒ.
    
    def __init__(self,owner,balance):
        self.owner = owner #ê³„ì¢Œì£¼
        self.__balance = balance or 0 #ì”ì•¡
    def bprint(self):
        return f'{self.owner}ë‹˜ì˜ ì”ê³ ëŠ” {self.__balance}ì› ì…ë‹ˆë‹¤.'
    def deposit(self, amount):
        self.__balance += amount
        return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì…ê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self.__balance}ì›"
    def withdraw(self, amount):
        if self.__balance >= amount:
            self.__balance -= amount
            return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì¶œê¸ˆ ë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self.__balance}ì›"
        return "ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤."
    def interest(self):
        total = self.__balance * (1 + self.interest_rate)
        return f"{self.interest_rate}%ì˜ ì´ìê°€ ë°˜ì˜ëœ ê¸ˆì•¡ì€ {total:.0f}ì› ì…ë‹ˆë‹¤."
        
        

acc3 = BankAccount("ì² ìˆ˜",50000)
acc4 = BankAccount("ì˜í¬",200000)
print(acc3.bprint())
print(acc4.bprint())
print(acc3.interest())
print(acc4.interest())
BankAccount.interest_rate = 0.03
print(acc3.interest())
print(acc4.interest())


#ê³µê²©í›„
#print(acc3.bprint())
#acc3.deposit(30000)
#acc3.withdraw(5500)
#
```

```python
class VendingMachine:
    def __init__(self):
        """ìë™íŒë§¤ê¸° ì´ˆê¸°í™”"""
        self.balance = 0 # ì‚¬ìš©ìì˜ ì”ì•¡
        self.items = { #ìƒí’ˆ ëª©ë¡ (ì´ë¦„:ê°€ê²©)
            "ì½œë¼" : 1500,
            "ì‚¬ì´ë‹¤" : 1400,
            "ë¬¼" : 1000
        }
    def insert_money(self, amount):
        """ì”ì•¡ ì¶©ì „"""
        self.balance += amount
        return f"ì”ì•¡ì´ {self.balance}ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤."
    def buy_item(self,item_name):
        """ìƒí’ˆ êµ¬ë§¤"""
        if item_name not in self.items:
            return f"{item_name}ì€(ëŠ”) íŒë§¤í•˜ì§€ ì•ŠëŠ” ìƒí’ˆì…ë‹ˆë‹¤."
        price = self.items[item_name]
        if self.balance < price:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤! ({item_name} ê°€ê²©: {price}ì›, í˜„ì¬ ì”ì•¡: {self.balance}ì›"
        self.balance -= price
        return f"{item_name}ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡: {self.balance}ì›"
    def refund(self):
        """ì”ì•¡ ë°˜í™˜"""
        amount = self.balance
        self.balance = 0
        return f"ì”ì•¡ {amount}ì›ì´ ë°˜í™˜ë˜ì—ˆìŠµë‹ˆë‹¤."
# ìë™ íŒë§¤ê¸° ê°ì²´ ìƒì„±
vm = VendingMachine()
# ì”ì•¡ ì¶©ì „
print(vm.insert_money(2000)) # ì”ì•¡ 2000ì› ì¦ê°€
# ìƒí’ˆ êµ¬ë§¤
print(vm.buy_item("ì½œë¼"))
print(vm.buy_item("ì‚¬ì´ë‹¤"))
# ì”ì•¡ ë°˜í™˜
print(vm.refund())
```

```python
import random

class Character: # ê´„í˜¸ëŠ” ìƒì†ë°›ì„ ë•Œ ì‚¬ìš©
    def __init__(self,name,hp,attack_p):
        """ìºë¦­í„° ì´ˆê¸°í™”"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self,opponent):
        """ìƒëŒ€ë°©ì„ ê³µê²©"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}ê°€ {opponent.name}ë¥¼ ê³µê²©í•˜ì—¬ {damage} ë°ë¯¸ì§€ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤.")
    def now_status(self):
        """í˜„ì¬ ì²´ë ¥ ìƒíƒœ í™•ì¸"""
        return f"{self.name} ì²´ë ¥: {self.hp}"
    def is_alive(self):
        """ìºë¦­í„° ìƒì¡´ ì—¬ë¶€ í™•ì¸"""
        return self.hp > 0
    

# íˆì–´ë¡œì™€ ëª¬ìŠ¤í„° ìƒì„±
player = Character(hp=50, name="ì˜ì›…", attack_p=15) #ìˆœì„œëŒ€ë¡œ ì•ˆ ë„£ê³  ì‹¶ì„ ë•Œ hp = ë¼ê³  ì§€ì •í•´ì¤˜ì•¼í•œë‹¤. ì²«ë²ˆì§¸ ì´ì™¸ì—ëŠ” ì§€ì • í•„ìˆ˜.
monster = Character("ê³¨ë ˜", 60, 9)


# ì „íˆ¬ ì§„í–‰
print("=====ì „íˆ¬ ì‹œì‘!!=====")
while player.is_alive and monster.is_alive():
    input("ê³µê²©í•˜ë ¤ë©´ ì—”í„°ë¥¼ ì¹˜ì„¸ìš”")
    player.attack(monster)
    if monster.is_alive():
        monster.attack(player)
    print(player.now_status(),' vs ', monster.now_status())
# ê²°ê³¼
if player.is_alive():
    print("ë‹¹ì‹ ì´ ìŠ¹ë¦¬í•˜ì˜€ìŠµë‹ˆë‹¤.ğŸ¥³")
else:
    print("ğŸ˜¢ğŸ˜¢ğŸ˜¢íŒ¨ë°°í•˜ì˜€ìŠµë‹ˆë‹¤... to be continue..?")
```

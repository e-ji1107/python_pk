```python
import pandas as pd
file_path = "./ref/서울시 자치구 (범죄예방 수사용) CCTV 설치현황_241231.xlsx"
CCTV = pd.read_excel(file_path)
CCTV
```

```python
cctv = CCTV.copy()
cctv = cctv.drop(index=[0,1,2,28,29])
cctv = cctv.drop(columns=["Unnamed: 0","Unnamed: 1"])
cctv
```

```python
data = cctv[['Unnamed: 2','Unnamed: 11']]
data.columns=["local","total"]
data
```

```python
# 'total' 값을 오름차순으로 정렬
data_sorted = data.sort_values(by="total", ascending=False)

data_sorted = data_sorted.reset_index(drop=True)
data_sorted = data_sorted.set_index('local')
data_sorted
```

```python
# OS에 사용가능한 , 설치된 폰트를 나열
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.font_manager as fm

# 리눅스용 폰트 경로 설정
plt.rc('font', family="NanumGothic")
```

```python
# plt.figure()
data_sorted['total'].sort_values().plot( kind="barh",grid=True,figsize=(7,7), ylabel="구")

plt.title("CCTV 구별 데이터")
plt.xlabel("CCTV 설치대 수")
plt.show()
```

```python
import pandas as pd

file_path2 = "./ref/5대+범죄+발생현황_20250307154603.csv"
ori = pd.read_csv(file_path2, encoding = "utf-8")
ori
```

```python
data1 = ori.copy()
data1 = data1.drop(index=0)
data1 = data1.drop(columns=["범죄별(1)","자치구별(1)","항목","시점"])
data1
```

```python
data1.columns = ["local","total"]
data1 = data1.reset_index(drop=True)
data1 = data1.sort_values(by="total", ascending=False)
data1 = data1.set_index("local")
data1
```

```python
total = pd.merge(data,data1, on='local')
total.columns=["local","count","incident"]
total = total.sort_values(by="count", ascending=False)
total = total.set_index('local')
total['count'] = total['count'].astype('int64')
total.dtypes
```

```python
import numpy as np
fp1 = np.polyfit(total['incident'],total['count'], 1)
incident_range = np.linspace(total['incident'].min(), total['incident'].max(), 100)

f1 = np.poly1d(fp1)
fx = np.linspace(10000,10000,1)
plt.figure(figsize=(9,9))
plt.plot(incident_range, f1(incident_range), ls='dashed', lw=3, color='g')
plt.scatter(total['incident'], total['count'], s= 30)
plt.xlabel('incident')
plt.ylabel('CCTV')
plt.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fp1 = np.polyfit(total['incident'], total['count'], 1)
f1 = np.poly1d(fp1)

fx = np.linspace(total['incident'].min(), total['incident'].max(), 100)

plt.figure(figsize=(14,8))
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')  # 초록 점선

# 오차 계산
total['err'] = np.abs(total['count'] - f1(total['incident']))

scatter = plt.scatter(total['incident'], total['count'], s=30, c=total['err'], cmap='viridis')
plt.xlabel('Incident')
plt.ylabel('CCTV')

for i in range(len(total['count'])):
    plt.text(total['incident'].iloc[i] * 0.95, total['count'].iloc[i] * 1.007, total.index[i], fontsize=15)


plt.colorbar(scatter, label='Error')
plt.grid(True)
plt.show()
```
